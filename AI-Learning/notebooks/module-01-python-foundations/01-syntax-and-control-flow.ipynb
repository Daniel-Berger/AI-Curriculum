## 1. Variables and Type Annotations

### Swift Mindset
In Swift, you distinguish between constants (`let`) and variables (`var`), and the compiler enforces immutability:

```swift
let name: String = "Daniel"   // immutable
var age: Int = 30              // mutable
age = 31                       // OK
// name = "Dan"               // Compile error!
```

### Python Approach
Python has **no `let` equivalent**. All variables are rebindable. Type hints exist but are **not enforced at runtime** - they're for documentation and static analysis tools like `mypy`.

````python
# Python variables - all are rebindable (like Swift's var)
name: str = "Daniel"       # type hint (not enforced!)
age: int = 30
age = 31                   # Fine
name = "Dan"               # Also fine - no compile error

print(f"Name: {name}, Age: {age}")
print(f"Type of name: {type(name)}")
print(f"Type of age: {type(age)}")
````

````python
# Python is dynamically typed - you can rebind to a different type!
x: int = 42
print(f"x = {x}, type = {type(x)}")

x = "now I'm a string"  # No error at runtime, but mypy would flag this
print(f"x = {x}, type = {type(x)}")

# Convention for "constants": use UPPER_CASE names
MAX_RETRIES: int = 3
API_BASE_URL: str = "https://api.example.com"
# Nothing stops you from changing these, but the convention signals intent
````

````python
# typing module provides more complex type hints (like Swift generics)
from typing import Optional, Final

# Final is Python's closest equivalent to Swift's `let` (but still not enforced at runtime)
PI: Final[float] = 3.14159

# Optional[X] is equivalent to Swift's X?
middle_name: Optional[str] = None  # like var middleName: String? = nil

# Modern Python 3.10+ alternative syntax
middle_name_v2: str | None = None  # equivalent to Optional[str]

print(f"PI = {PI}")
print(f"middle_name = {middle_name}")
````

> **Key Takeaway**: Python's type hints are like writing Swift with every type annotation being a comment. They help readability and tooling but offer zero runtime safety. Use `mypy` in your projects to get Swift-like static checking.

## 2. Strings

### Swift Mindset
```swift
let name = "Daniel"
let greeting = "Hello, \(name)!"           // interpolation
let multi = """
    Line 1
    Line 2
    """                                     // multi-line
let count = name.count                       // character count
```

### Python Approach
Python strings are simpler in some ways (no Character vs String distinction) but offer more formatting options.

````python
# String basics
name = "Daniel"

# f-strings (Python 3.6+) - equivalent to Swift's \() interpolation
greeting = f"Hello, {name}!"
print(greeting)

# You can put ANY expression inside {}
print(f"2 + 3 = {2 + 3}")
print(f"Name uppercased: {name.upper()}")
print(f"Name length: {len(name)}")

# Formatting numbers (no equivalent in Swift interpolation)
pi = 3.14159
print(f"Pi to 2 decimal places: {pi:.2f}")
print(f"Large number with commas: {1_000_000:,}")
print(f"Percentage: {0.856:.1%}")
````

````python
# Multi-line strings (triple quotes - like Swift's multi-line string literals)
multi = """Line 1
Line 2
Line 3"""
print(multi)
print()

# Common string methods (Swift equivalents in comments)
s = "  Hello, World!  "
print(f"strip:      '{s.strip()}'"  )      # trimmingCharacters(in: .whitespaces)
print(f"lower:      '{s.lower()}'"  )      # lowercased()
print(f"upper:      '{s.upper()}'"  )      # uppercased()
print(f"split:      {s.split(',')}")       # components(separatedBy: ",")
print(f"replace:    '{s.replace('World', 'Python')}'"  )  # replacingOccurrences
print(f"startswith: {s.strip().startswith('Hello')}")     # hasPrefix
print(f"endswith:   {s.strip().endswith('!')}")           # hasSuffix
print(f"find:       {s.find('World')}")    # range(of:) - returns index or -1
print(f"in:         {'World' in s}")       # contains()
````

````python
# String slicing - Python's killer feature (no direct Swift equivalent)
s = "Hello, World!"

print(f"s[0]     = '{s[0]}'")       # First character
print(f"s[-1]    = '{s[-1]}'")      # Last character (negative indexing!)
print(f"s[0:5]   = '{s[0:5]}'")    # First 5 chars (prefix(5))
print(f"s[7:]    = '{s[7:]}'")     # From index 7 to end
print(f"s[:5]    = '{s[:5]}'")     # Same as s[0:5]
print(f"s[::2]   = '{s[::2]}'")   # Every 2nd character
print(f"s[::-1]  = '{s[::-1]}'")  # Reversed! (String(s.reversed()))
````

## 4. Booleans, None, and Truthiness

### Swift Mindset
```swift
let flag: Bool = true
var name: String? = nil     // Optional
if let name = name { ... }  // Optional binding
```

### Python Approach
Python has a concept of **truthiness** that Swift deliberately avoids. Many values can be implicitly treated as booleans.

````python
# Booleans
is_active: bool = True   # Note: True/False are capitalized (not true/false)
is_done: bool = False

# Logical operators use English words (not && || !)
print(f"True and False = {True and False}")   # Swift: &&
print(f"True or False  = {True or False}")    # Swift: ||
print(f"not True       = {not True}")         # Swift: !

# None - Python's nil
value = None
print(f"value is None: {value is None}")    # ALWAYS use 'is' to check None
print(f"value == None: {value == None}")    # Works but not idiomatic
````

````python
# Truthiness - THIS IS A BIG DIFFERENCE FROM SWIFT
# In Swift, only Bool can be used in conditions. In Python, many values are "falsy".

# Falsy values (evaluate to False):
falsy_values = [False, None, 0, 0.0, "", [], {}, set(), tuple()]

print("Falsy values:")
for val in falsy_values:
    print(f"  bool({val!r:10s}) = {bool(val)}")

print("\nTruthy values:")
truthy_values = [True, 1, -1, 3.14, "hello", [1], {"a": 1}]
for val in truthy_values:
    print(f"  bool({str(val):10s}) = {bool(val)}")
````

````python
# Practical use of truthiness (very Pythonic)

# Instead of: if len(my_list) > 0:
my_list = [1, 2, 3]
if my_list:  # Truthy because non-empty
    print("List has items")

# Instead of: if name != "" and name is not None:
name = "Daniel"
if name:  # Truthy because non-empty string
    print(f"Hello, {name}")

# Guard-like pattern using 'or' (similar to Swift's ??)
username = None
display_name = username or "Anonymous"  # like: username ?? "Anonymous"
print(f"Display name: {display_name}")

# Walrus operator := (Python 3.8+) - similar to if-let
data = [1, 2, 3, 4, 5]
if (n := len(data)) > 3:  # assign and test in one expression
    print(f"List is long: {n} items")
````

## 5. If / Else / Elif

### Swift Mindset
```swift
if temperature > 30 {
    print("Hot")
} else if temperature > 20 {
    print("Nice")
} else {
    print("Cold")
}
// Ternary: let status = age >= 18 ? "adult" : "minor"
```

### Python Approach
No braces. Indentation IS the syntax. `elif` instead of `else if`.

````python
temperature = 25

# Standard if/elif/else
if temperature > 30:
    print("Hot")
elif temperature > 20:    # Note: elif, not "else if"
    print("Nice")
elif temperature > 10:
    print("Cool")
else:
    print("Cold")

# Ternary expression (reads like English)
age = 25
status = "adult" if age >= 18 else "minor"  # Swift: age >= 18 ? "adult" : "minor"
print(f"Status: {status}")

# Chained comparisons (unique to Python!)
x = 15
if 10 < x < 20:           # Swift: x > 10 && x < 20
    print(f"{x} is between 10 and 20")

# Identity vs equality
a = [1, 2, 3]
b = [1, 2, 3]
print(f"a == b: {a == b}")   # True (value equality, like Swift's ==)
print(f"a is b: {a is b}")   # False (identity, like Swift's ===)
````

## 6. For Loops

### Swift Mindset
```swift
for i in 0..<5 { print(i) }                    // 0,1,2,3,4
for i in stride(from: 0, to: 10, by: 2) { }    // 0,2,4,6,8
for (index, value) in array.enumerated() { }    // index + value
for item in collection { }                       // iterate items
```

````python
# range() - Python's equivalent of Swift ranges
# range(stop)          ->  0..<stop
# range(start, stop)   ->  start..<stop
# range(start, stop, step) -> stride(from:to:by:)

print("range(5):")
for i in range(5):         # 0..<5
    print(f"  {i}", end=" ")
print()

print("range(2, 8):")
for i in range(2, 8):      # 2..<8
    print(f"  {i}", end=" ")
print()

print("range(0, 10, 2):")
for i in range(0, 10, 2):  # stride(from: 0, to: 10, by: 2)
    print(f"  {i}", end=" ")
print()

print("range(10, 0, -1):")
for i in range(10, 0, -1): # Counting down
    print(f"  {i}", end=" ")
print()
````

````python
# Iterating collections
fruits = ["apple", "banana", "cherry"]

# Direct iteration (like Swift: for fruit in fruits)
for fruit in fruits:
    print(fruit)

print()

# enumerate() - like Swift's .enumerated()
for index, fruit in enumerate(fruits):
    print(f"{index}: {fruit}")

print()

# enumerate with custom start index
for index, fruit in enumerate(fruits, start=1):
    print(f"{index}. {fruit}")
````

````python
# zip() - iterate multiple sequences in parallel (like Swift's zip())
names = ["Alice", "Bob", "Charlie"]
scores = [95, 87, 92]

for name, score in zip(names, scores):
    print(f"{name}: {score}")

print()

# Iterating dictionaries
ages = {"Alice": 30, "Bob": 25, "Charlie": 35}

for name in ages:                    # Iterates keys by default
    print(f"{name}", end=" ")
print()

for name, age in ages.items():       # Key-value pairs
    print(f"{name} is {age}")

# for-else (unique to Python! no Swift equivalent)
# The else block runs if the loop completes WITHOUT a break
numbers = [1, 3, 5, 7, 9]
for n in numbers:
    if n % 2 == 0:
        print(f"Found even number: {n}")
        break
else:
    print("No even numbers found!")  # This runs because we never broke
````

## 7. While Loops, Break, and Continue

These work almost identically to Swift, just without braces.

````python
# While loop
count = 0
while count < 5:
    print(count, end=" ")
    count += 1  # Note: NO ++ operator in Python!
print()

# break and continue work the same as Swift
for i in range(10):
    if i == 3:
        continue  # Skip 3
    if i == 7:
        break     # Stop at 7
    print(i, end=" ")
print()

# Infinite loop pattern (like Swift's while true)
import random
attempts = 0
while True:
    attempts += 1
    roll = random.randint(1, 6)
    if roll == 6:
        print(f"Rolled a 6 after {attempts} attempts!")
        break
````

## 8. Match/Case (Python 3.10+)

### Swift Mindset
```swift
switch statusCode {
case 200:
    print("OK")
case 404:
    print("Not Found")
case 400..<500:
    print("Client error")
default:
    print("Other")
}
```

### Python's match/case
Added in Python 3.10, this is Python's answer to Swift's powerful `switch`. It supports pattern matching similar to Swift.

````python
# Basic match/case
def describe_status(code: int) -> str:
    match code:
        case 200:
            return "OK"
        case 301:
            return "Moved Permanently"
        case 404:
            return "Not Found"
        case 500:
            return "Server Error"
        case _:          # underscore is the wildcard (like Swift's default)
            return f"Unknown status: {code}"

for code in [200, 301, 404, 418, 500]:
    print(f"{code}: {describe_status(code)}")
````

````python
# Pattern matching with guards and OR patterns
def classify_number(n: int) -> str:
    match n:
        case 0:
            return "zero"
        case 1 | 2 | 3:                    # OR pattern (Swift: case 1, 2, 3)
            return "small positive"
        case n if n < 0:                    # Guard (Swift: case let n where n < 0)
            return "negative"
        case n if n > 100:                  # Guard
            return "large"
        case _:
            return "other positive"

for num in [0, 2, -5, 150, 42]:
    print(f"{num}: {classify_number(num)}")
````

````python
# Structural pattern matching (destructuring - like Swift's tuple matching)
def describe_point(point: tuple) -> str:
    match point:
        case (0, 0):
            return "Origin"
        case (x, 0):
            return f"On x-axis at x={x}"
        case (0, y):
            return f"On y-axis at y={y}"
        case (x, y):
            return f"Point at ({x}, {y})"

points = [(0, 0), (5, 0), (0, 3), (2, 7)]
for p in points:
    print(f"{p} -> {describe_point(p)}")
````

## 9. Comprehensions - Python's Superpower

Comprehensions are one of Python's most distinctive features. They replace common `map`, `filter`, and `reduce` patterns with concise, readable syntax.

### Swift Comparison
```swift
// Swift
let squares = (0..<10).map { $0 * $0 }
let evens = numbers.filter { $0 % 2 == 0 }
let dict = zip(keys, values).reduce(into: [:]) { $0[$1.0] = $1.1 }
```

````python
# LIST COMPREHENSIONS
# Syntax: [expression for item in iterable if condition]

# Basic - equivalent to Swift's .map
squares = [x ** 2 for x in range(10)]
print(f"Squares: {squares}")

# With filter - equivalent to Swift's .filter then .map
even_squares = [x ** 2 for x in range(10) if x % 2 == 0]
print(f"Even squares: {even_squares}")

# With transformation
words = ["hello", "WORLD", "Python"]
lower_words = [w.lower() for w in words]
print(f"Lowered: {lower_words}")

# Nested loops (like flatMap)
pairs = [(x, y) for x in range(3) for y in range(3)]
print(f"Pairs: {pairs}")

# If-else in comprehension (note: different position!)
labels = ["even" if x % 2 == 0 else "odd" for x in range(6)]
print(f"Labels: {labels}")
````

````python
# DICTIONARY COMPREHENSIONS
# Syntax: {key_expr: value_expr for item in iterable if condition}

# Create a dict from a list
names = ["Alice", "Bob", "Charlie"]
name_lengths = {name: len(name) for name in names}
print(f"Name lengths: {name_lengths}")

# Invert a dictionary (swap keys and values)
original = {"a": 1, "b": 2, "c": 3}
inverted = {v: k for k, v in original.items()}
print(f"Inverted: {inverted}")

# Filter a dictionary
scores = {"Alice": 95, "Bob": 67, "Charlie": 82, "Diana": 91}
high_scores = {name: score for name, score in scores.items() if score >= 80}
print(f"High scores: {high_scores}")
````

````python
# SET COMPREHENSIONS
# Syntax: {expression for item in iterable if condition}

sentence = "the quick brown fox jumps over the lazy dog"
unique_lengths = {len(word) for word in sentence.split()}
print(f"Unique word lengths: {sorted(unique_lengths)}")

# GENERATOR EXPRESSIONS (lazy evaluation - like Swift's lazy sequences)
# Syntax: (expression for item in iterable if condition)
# Uses () instead of [] - doesn't create the whole list in memory

# This creates all million numbers in memory:
# big_list = [x ** 2 for x in range(1_000_000)]

# This generates them one at a time (memory efficient):
big_gen = (x ** 2 for x in range(1_000_000))
print(f"Generator type: {type(big_gen)}")
print(f"Sum of first million squares: {sum(big_gen):,}")
````

## 10. Putting It All Together - Side by Side Comparison

Let's see a complete example written in both Swift style and Pythonic style.

````python
# "Swift-like" Python (how you might write it at first)
def process_grades_swift_style(grades: dict) -> None:
    passing_students = []
    for name in grades:
        grade = grades[name]
        if grade >= 70:
            passing_students.append(name)
    
    total = 0
    for name in passing_students:
        total = total + grades[name]
    
    if len(passing_students) > 0:
        average = total / len(passing_students)
        print(f"Average passing grade: {average:.1f}")
    else:
        print("No passing students")

# Pythonic version
def process_grades_pythonic(grades: dict) -> None:
    passing = {name: grade for name, grade in grades.items() if grade >= 70}
    
    if passing:
        average = sum(passing.values()) / len(passing)
        print(f"Average passing grade: {average:.1f}")
    else:
        print("No passing students")

grades = {"Alice": 95, "Bob": 65, "Charlie": 82, "Diana": 91, "Eve": 58}
process_grades_swift_style(grades)
process_grades_pythonic(grades)
````

````python
# SOLUTION - Exercise 3

def word_frequencies(text: str) -> dict[str, int]:
    freq: dict[str, int] = {}
    for word in text.lower().split():
        freq[word] = freq.get(word, 0) + 1  # .get() is like dict[key, default:] in Swift
    return freq

def top_n_words(freq: dict[str, int], n: int = 3) -> list[tuple[str, int]]:
    # sorted() with key function, reverse for descending
    sorted_words = sorted(freq.items(), key=lambda pair: pair[1], reverse=True)
    return sorted_words[:n]

sample = "the cat sat on the mat the cat liked the mat"
freq = word_frequencies(sample)
print(f"Frequencies: {freq}")
print(f"Top 3: {top_n_words(freq)}")

assert freq["the"] == 4
assert freq["cat"] == 2
print("All tests passed!")
````

````python
# SOLUTION - Exercise 5

def parse_command(command: str) -> str:
    parts = command.strip().split()
    
    match parts:
        case ["quit" | "exit"]:
            return "Goodbye!"
        case ["hello", name]:
            return f"Hello, {name}!"
        case ["add", x, y]:
            try:
                return f"Result: {int(x) + int(y)}"
            except ValueError:
                return f"Error: '{x}' and '{y}' must be numbers"
        case ["repeat", n, word]:
            try:
                return word * int(n)
            except ValueError:
                return f"Error: '{n}' must be a number"
        case _:
            return "Unknown command"

# Tests
print(parse_command("quit"))           # Goodbye!
print(parse_command("exit"))           # Goodbye!
print(parse_command("hello Daniel"))   # Hello, Daniel!
print(parse_command("add 3 5"))        # Result: 8
print(parse_command("repeat 3 ha"))    # hahaha
print(parse_command("unknown stuff"))  # Unknown command

assert parse_command("quit") == "Goodbye!"
assert parse_command("hello Daniel") == "Hello, Daniel!"
assert parse_command("add 3 5") == "Result: 8"
assert parse_command("repeat 3 ha") == "hahaha"
print("\nAll tests passed!")
````

## Key Takeaways
